```python
def check_row(row):
    ouput = True
    for i, j in enumerate(row):
        if j in row[i + 1:]:
            return False
    return True

def check_row_list(row_list):
    for i in row_list:
        if check_row(i) == False:
            return False
    return True

def check_column_list(sudoku_list):
    column_list = list()
    for i in range(9):
        column_list.append([])
    for row in sudoku_list:
        for i, j in enumerate(row):
            column_list[i].append(j)
    for i in column_list:
        if check_row(i) == False:
            return False
    return True

def check_square_list(sudoku_list):
    square_list = list()
    for i in range(9):
        square_list.append([])
    center_list = [[1, 1], [1, 4], [1, 7], [4, 1], [4, 4], [4, 7], [7, 1], [7, 4], [7, 7]]
    for i, j in enumerate(center_list):
        x, y = j
        square_list[i].append(sudoku_list[x - 1][y - 1])
        square_list[i].append(sudoku_list[x - 1][y])
        square_list[i].append(sudoku_list[x - 1][y + 1])
        square_list[i].append(sudoku_list[x][y - 1])
        square_list[i].append(sudoku_list[x][y])
        square_list[i].append(sudoku_list[x][y + 1])
        square_list[i].append(sudoku_list[x + 1][y - 1])
        square_list[i].append(sudoku_list[x + 1][y])
        square_list[i].append(sudoku_list[x + 1][y + 1])
    for i in square_list:
        if check_row(i) == False:
            return False
    return True
                          
def main():
    sudoku_list = list()
    for i in range(9):
        sudoku_list.append(input().split())
    if check_row_list(sudoku_list) == False:
        return False
    if check_column_list(sudoku_list) == False:
        return False
    if check_square_list(sudoku_list) == False:
        return False
    return True
        

N = int(input())        
for i in range(N):
    if i == N - 1:
        output = main()
        if output == False:
            print('Invalid')
        else:
            print('Valid')
    else:
        output = main()
        if output == False:
            print('Invalid')
        else:
            print('Valid')
        input()
```
